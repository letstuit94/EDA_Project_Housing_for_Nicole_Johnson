Importing the Data from PostgreSQL to join tables and create our dataframe to work on

```python
import pandas as pd
import psycopg2
```

```python
#load connection data from .env file into notebook
import os
from dotenv import load_dotenv

load_dotenv()

DATABASE = os.getenv("DATABASE")
USER_DB = os.getenv("USER_DB")
PASSWORD = os.getenv("PASSWORD")
HOST = os.getenv("HOST")
PORT = os.getenv("PORT")
```

```python
print(USER_DB)
```

```python
# Create connection object conn
conn = psycopg2.connect(
    database=DATABASE, user=USER_DB, password=PASSWORD, host=HOST, port=PORT
)
```

```python

cur = conn.cursor()
```

```python
query_string = "select * from eda.king_county_house_details hd full outer join eda.king_county_house_sales hs on hd.id=hs.house_id"
df_housingdata = pd.read_sql(query_string, conn)
```
